```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import rand_score
import matplotlib.pyplot as plt
import numpy as np
```

```python
df =pd.read_csv ('seeds_dataset.txt', sep='\s+',header=None)
df
```

```python
target = df[7]
df = df.drop(columns = 7)

print (df)
```

```python
x = df.to_numpy()

print (x)

aggloclust=AgglomerativeClustering(n_clusters=3).fit(x)
print (aggloclust)

kmeans = KMeans(n_clusters=3, random_state=0).fit(x)
print(kmeans)

pca = PCA(n_components=7)
x = pca.fit_transform(x)

print (x)

agglolabels = aggloclust.labels_

plt.scatter(x[:,0], x[:,1], c=agglolabels)
plt.show()

kmeanlabels = kmeans.labels_

plt.scatter(x[:,0], x[:,1], c=kmeanlabels)
plt.show()

rand_score(target, kmeanlabels)

rand_score(target, agglolabels)
```

```python
r1 =rand_score(target, kmeanlabels)

r2 = rand_score(target, agglolabels)

print (r1)

print (r2)
```
