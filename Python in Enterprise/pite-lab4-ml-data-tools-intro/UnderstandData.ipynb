## Modify your dataset

### Add a column to the dataset. 

If want to add a new column in our dataset, as we are doing any calculations or extracting some information from the dataset, 
and if you want to save it a new column... 

This can be done by the following code by taking a case where we have added all integer values of all columns.

```python
# For example, if we want to add a column let say "total_values",
# that means if you want to add all the integer value of that particular
# row and get total answer in the new column "total_values".
# first we will extract the columns which have integer values.
cols = data.columns
 
# it will print the list of column names.
print(cols)
 
# we will take that columns which have integer values.
cols = cols[1:5]
# we will save it in the new dataframe variable
data1 = data[cols]
 
# now adding new column "total_values" to dataframe data.
data["total_values"]=data1[cols].sum(axis=1)
 
# here axis=1 means you are working in rows,
# whereas axis=0 means you are working in columns.


data.head()
```

### Strip plot 

A strip plot is a graphical data anlysis technique for summarizing a univariate data set. 

That is, a strip plot is simply a plot of the sorted response values along one axis. 

The strip plot is an alternative to a histogram or a density plot. 

It is typically used for small data sets (histograms and density plots are typically preferred for larger data sets).

```python
sns.stripplot(x='class',y='sepal-length',data=data,jitter=True,edgecolor='gray')
```

Print out the most important information of Iris dataset.

##### Hint for correlation: use values from 'total values' column from the output of data.corr. If you have removed 'total_values' column you may add this once again. 
See [Add a column to the dataset.](#Add-a-column-to-the-dataset.)


```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
data = pd.read_csv(url, names=names)
data.head(10)

cols = data.columns
cols = cols[0:4]
data1 = data[cols]
data["total_values"]=data1[cols].sum(axis=1)

print("Attribute Information:")
for col in names:
    print("->",col,end='')
    if col == 'class':
        print(":")
        for each in data[col].unique():
            print("\t",each)
    else:
        print(" in cm")

print("\nNumber of instances:",data.shape[0])


print("\nSummary Statistics:")
sum_dic={}
for col in names[0:4]:  
    sum_dic[col]={"Min": data[col].min(),"Max": data[col].max(),"Mean": data[col].mean(),"SD": data[col].std(),"Class Correlation": data[col].corr(data["total_values"])}

new_data=pd.DataFrame(sum_dic)
result=new_data.transpose()
print(result)

print("\nClass Distribution:",end='')
class_distribution = data["class"].value_counts(normalize=True)
print(f"{class_distribution[0]*100}% for each of {len(class_distribution)} classes")

```
